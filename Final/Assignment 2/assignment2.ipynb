# Object detection using YOLO

```python
#This command is used to detect objects in the giraffe.jpg image using the YOLOv3 model. 

#The "cfg/yolov3.cfg" part specifies the configuration file for the YOLOv3 model. 
#This file contains the architecture of the neural network that YOLOv3 uses to detect objects.

#The "yolov3.weights" part specifies the pre-trained weights for the YOLOv3 model,

!./darknet detect cfg/yolov3.cfg yolov3.weights data/giraffe.jpg
```

```python
#imports the OpenCV library, Matplotlib library
import cv2                      
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread("predictions.jpg")  #read image
height, width = image.shape[:2]       #extracts the height and width of the image
resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC) #resize the image to three times its original size

fig = plt.gcf()
fig.set_size_inches(18, 10) #set the size of the figure to 18 inches in width and 10 inches in height
plt.axis("off")
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)) #display the resized image and convert the color space from BGR to RGB
plt.show() #display the image
```
