```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

Step function:

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
data = np.arange(-10, 10, 0.1)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y, 'r')
plt.title('Step Function')
plt.ylabel('Step(X)')
plt.show()
```

Sigmoid Function:

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

data = np.arange(-10, 10, 0.1)

plt.figure(figsize=(10,10))
plt.title('Sigmoid Function')
plt.plot(data, sigmoid(data), 'r')
plt.show()
```

```python
test_data = np.arange(-10, 10, 0.1)

def derivative_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

plt.figure(figsize=(10,10))
plt.title('Derivative of Sigmoid Function')
plt.plot(test_data, derivative_sigmoid(test_data), 'r')
plt.show()
```

Tanh function:

```python
test_data = np.arange(-10, 10, 0.1)

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.figure(figsize=(10,10))
plt.title('Tanh Function')
plt.plot(test_data, tanh(test_data), 'r')
plt.show()
```

```python
def tanh_derivative(x):
    return 1 - tanh(x)**2
```

```python
plt.figure(figsize=(10,10))
plt.title('Derivative of Tanh Function')
plt.plot(test_data, tanh_derivative(test_data), 'r')
plt.show()
```

ReLU Function:

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.figure(figsize=(10,10))
plt.title('ReLU Function')
plt.plot(test_data, relu(test_data), 'r')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Relu Problem')
plt.plot(test_data, np.where(test_data > 0, 1, 0), 'r')
plt.show()
```

```python
def leaky_relu_derivative(x):
    return np.where(x > 0, 1, 0.01)
```

```python
plt.figure(figsize=(10,10))
plt.title('Derivative of Leaky ReLU Function')
plt.plot(test_data, leaky_relu_derivative(test_data), 'r')
plt.show()
```

Exponential Linear Units (ELUs) Function:

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha*(np.exp(x)-1))

plt.figure(figsize=(10,10))
plt.title('ELU Function')
plt.plot(test_data, elu(test_data), 'r')
plt.show()
```

Scaled Exponential Linear Units (SELUs) Function:

```python
 def selu(x, scale=1.0507, alpha=1.6732):
    return np.where(x >= 0, scale*x, scale*(alpha*np.exp(x)-alpha))
```

```python
plt.figure(figsize=(10,10))
plt.title('SELU Function')
plt.plot(test_data, selu(test_data), 'r')
plt.show()
```
